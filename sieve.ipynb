```python
import os
from pathlib import Path
from my_utils import sieve
from tqdm.notebook import tqdm
from pydub import AudioSegment
from concurrent.futures import ProcessPoolExecutor
```

```python
path = Path(r'D:\audio\ds\clips')
save_path = Path(r'C:\audio_classificator\dataset\chunks\human')

file_list = path.iterdir()
```

```python
def main():

    tasks = []

    for file in file_list:
        audio_path = path / file
        tasks.append((audio_path, save_path))
    
    with ProcessPoolExecutor() as executor:
        list(tqdm(executor.map(sieve, tasks), total=len(tasks)))
```

```python
if __name__ == '__main__':
    main()
```
